```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
# Load dataset
df = pd.read_csv("csv_files/myntra_products_catalog.csv")

df.rename(columns={"Price (INR)": "Price_INR"}, inplace=True)
```

```python
## Price Distribution
plt.figure(figsize=(8, 5))
sn.histplot(df['Price_INR'], bins=50, kde=True)
plt.title("Price Distribution")
plt.xlabel("Price (INR)")
plt.show()
```

```python
## Top 10 Product Brands by Count

top_brands = df['ProductBrand'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sn.barplot(x=top_brands.values, y=top_brands.index)
plt.title("Top 10 Product Brands")
plt.xlabel("Count")
plt.ylabel("Product Brand")
plt.show()
```

```python
## Price Distribution by Gender

plt.figure(figsize=(8, 6))
sn.boxplot(data=df, x='Gender', y='Price_INR')
plt.title("Price Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Price (INR)")
plt.show()
```

```python
## Correlation Heatmap

plt.figure(figsize=(6, 4))
sn.heatmap(df[['Price_INR', 'NumImages']].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
